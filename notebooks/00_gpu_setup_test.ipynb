- MPS доступна: **True**

- Тензор на MPS создан успешно


```python
import torch

print("PyTorch версия:", torch.__version__)
print("MPS доступна (GPU Apple Silicon):", torch.backends.mps.is_available())
print("Текущее устройство:", "mps" if torch.backends.mps.is_available() else "cpu")

# Дополнительно — проверь, сколько памяти доступно на MPS
if torch.backends.mps.is_available():
    print("MPS backend работает!")
    # Простой тест тензора на MPS
    x = torch.randn(1000, 1000, device="mps")
    print("Тензор успешно создан на MPS!")
else:
    print("Работаем на CPU.")
```
